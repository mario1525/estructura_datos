```python
codigoCifrado=input("escribe el codigo texto: ")

a="ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
b="abcdefghijklmnñopqrstuwxyz"
c="0123456789" 

k=int(input("Valor de desplazamiento: "))
cifrad=""
    
for d in codigoCifrado:
    if d in a:
        cifrad += a[(a.index(d)+k)%(len(a))]
    if d in b:
        cifrad += b[(b.index(d)+k)%(len(b))]
    if d in c:
        cifrad += a[(c.index(d)+k)%(len(c))]
        
    else:
        cifrad+=a
        cifrad+=b
        cifrad+=c

print("Texto cifrado: ",cifrad)
 
texto=input("Mensaje: ")

caracts='ABCDEFGHIJKLMNÑOPQRSTUVWXYZabcdefghijklmnñopqrstuwxyz0123456789'

for key in range(len(caracts)):
    traducido=''

    for elem in codigoCifrado:
        if elem in caracts:
            elemIndex=caracts.find(elem)
            tradIndex=elemIndex-key

            if tradIndex<0:
                tradIndex=tradIndex+len(caracts)
            traducido=traducido+caracts[tradIndex]
        else:
            traducido=traducido+elem

        if len(traducido)==len(codigoCifrado):
            print('key #%s: %s' % (key, traducido))
```
